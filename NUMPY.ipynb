Array Reshaping ---> Reshape, Ravel, Flatten

```python
# Reshape

arr = np.array([1,2,3,4,5,6])
arr
reshaped = arr.reshape(3,2)    # 3->rows & 2->columns
reshaped
```

```python
# Ravel

# used to convert array into 1D array
# returns the view of the array

ravel = reshaped.ravel()
ravel
```

```python
ravel[0] = 20
print(ravel)
print(reshaped)
```

```python
# Flatten

# it is also used to convert array into 1D array
# difference is ---> it returns the copy of the array

flat = reshaped.flatten()
flat
flat[0] = 1
print(reshaped)
```

```python
# whenever we convert array into ravel() the changes are reflected in the original array
# and in flatten the changes are not reflected into array, the independent array is created
```
